Feature Engineering

```python
#feature extraction with RFE
import warnings
warnings.filterwarnings("ignore")
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

x = data.iloc[:,1:]
y = data['Bankrupt?']

model = LogisticRegression(max_iter=1000)
rfe = RFE(model, n_features_to_select=10)
fit = rfe.fit(x,y)

print(fit.n_features_)
print(fit.support_)
print(fit.ranking_)
```

```python
rfe_selected_feature = x.columns[fit.support_]
```

```python
#feature extraction with Extra Erees Classifier

from sklearn.ensemble import ExtraTreesClassifier

x = data.iloc[:,1:]
y = data['Bankrupt?']

model = ExtraTreesClassifier()
model.fit(x,y)

print(model.feature_importances_)
```

```python
data[rfe_selected_feature].corr()
```

```python
#extract important features using correlation matrix
correlation_matrix = data.corr()
correlation_matrix['Bankrupt?'].sort_values(ascending=False)
```

```python
important_features = correlation_matrix['Bankrupt?'][correlation_matrix['Bankrupt?']>=0].index
df = data[important_features]
```

```python
df.shape
```

```python
x = df.drop('Bankrupt?',axis=1)
y = df['Bankrupt?']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Model Building

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = LogisticRegression()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

logistic_accuracy = accuracy_score(Y_test, predictions)
logistic_conf_matrix = confusion_matrix(Y_test, predictions)
logistic_classification_report = classification_report(Y_test, predictions)

print(f"Accuracy: {logistic_accuracy}")
print(f"Confusion Matrix:\n{logistic_conf_matrix}")
print(f"Classification Report:\n{logistic_classification_report}")
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

dtree_accuracy = accuracy_score(Y_test, predictions)
dtree_conf_matrix = confusion_matrix(Y_test, predictions)
dtree_classification_report = classification_report(Y_test, predictions)

print(f"Accuracy: {dtree_accuracy}")
print(f"Confusion Matrix:\n{dtree_conf_matrix}")
print(f"Classification Report:\n{dtree_classification_report}")
```

```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

model = DecisionTreeClassifier()
model.fit(X_train,Y_train)

y_prob = model.predict_proba(X_test)

fpr, tpr, thresholds = roc_curve(Y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
#random forest 
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

rforest_accuracy = accuracy_score(Y_test, predictions)
rforest_conf_matrix = confusion_matrix(Y_test, predictions)
rforest_classification_report = classification_report(Y_test, predictions)

print(f"Accuracy: {rforest_accuracy}")
print(f"Confusion Matrix:\n{rforest_conf_matrix}")
print(f"Classification Report:\n{rforest_classification_report}")
```

```python
#support vector classifier
from sklearn.svm import SVC

model = SVC()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

svc_accuracy = accuracy_score(Y_test, predictions)
svc_conf_matrix = confusion_matrix(Y_test, predictions)
svc_classification_report = classification_report(Y_test, predictions)

print(f"Accuracy: {svc_accuracy}")
print(f"Confusion Matrix:\n{svc_conf_matrix}")
print(f"Classification Report:\n{svc_classification_report}")
```

```python
#Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

gboost_accuracy = accuracy_score(Y_test, predictions)
gboost_conf_matrix = confusion_matrix(Y_test, predictions)
gboost_classification_report = classification_report(Y_test, predictions)

print(f"Accuracy: {gboost_accuracy}")
print(f"Confusion Matrix:\n{gboost_conf_matrix}")
print(f"Classification Report:\n{gboost_classification_report}")
```

```python
#KNN 
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

knn_accuracy = accuracy_score(Y_test, predictions)
knn_conf_matrix = confusion_matrix(Y_test, predictions)
knn_classification_report = classification_report(Y_test, predictions)

print(f"Accuracy: {knn_accuracy}")
print(f"Confusion Matrix:\n{knn_conf_matrix}")
print(f"Classification Report:\n{knn_classification_report}")
```

```python
#Adaboost classifier
from sklearn.ensemble import AdaBoostClassifier

model = AdaBoostClassifier()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

adaboost_accuracy = accuracy_score(Y_test, predictions)
adaboost_conf_matrix = confusion_matrix(Y_test, predictions)
adaboost_classification_report = classification_report(Y_test, predictions)

print(f"Accuracy: {adaboost_accuracy}")
print(f"Confusion Matrix:\n{adaboost_conf_matrix}")
print(f"Classification Report:\n{adaboost_classification_report}")
```

```python
# Gaussian Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

gaussian_accuracy = accuracy_score(Y_test, predictions)
gaussian_conf_matrix = confusion_matrix(Y_test, predictions)
gaussian_classification_report = classification_report(Y_test, predictions)

print(f"Accuracy: {gaussian_accuracy}")
print(f"Confusion Matrix:\n{gaussian_conf_matrix}")
print(f"Classification Report:\n{gaussian_classification_report}")
```

# Model Validation

```python
#Evaluate using train test split method

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

data = pd.read_csv("Bankruptcy Prediction.csv")
x = data.drop("Bankrupt?",axis=1)
y = data['Bankrupt?']

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = RandomForestClassifier()
model.fit(X_train,Y_train)
prediction = model.predict(X_test)
accuracy = accuracy_score(Y_test,prediction)

print(f"Accuracy : {accuracy}")
```

```python
#evaluating using cross validating
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

data = pd.read_csv("Bankruptcy Prediction.csv")
x = data.drop("Bankrupt?",axis=1)
y = data['Bankrupt?']


model = RandomForestClassifier()
cv_score = cross_val_score(model,x,y,cv=5)

mean_accuracy = np.mean(cv_score)
std_accuracy = np.std(cv_score)

print(f'Mean Accuracy: {mean_accuracy}')
print(f'Standard Deviation: {std_accuracy}')
```

# Deployment

```python
import joblib
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

data = pd.read_csv("Bankruptcy Prediction.csv")
x = data.drop("Bankrupt?",axis=1)
y = data['Bankrupt?']

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = RandomForestClassifier()
model = model.fit(X_train,Y_train)

joblib.dump(model,'Bankruptcy_train_model.joblib')
```

```python
def predict_model(new_data):
    model = joblib.load('Bankruptcy_train_model.joblib')
    prediction = model.predict(new_data)
    return prediction
```
